```python
# 실습 #7

def wc():
    input_wc = input(">> ")
    input_wc_array = input_wc.strip().split(" ")
    if len(input_wc_array) != 2:
        print("wc file 과 같이 입력해주세요.")
    elif input_wc_array[0] != "wc":
        print(input_wc_array[0], "은 확인할 수 없는 명령어 입니다.")
    else:
        filePath = input_wc_array[1]
        f = open(filePath, 'r', encoding='UTF8')

        lineCount = 0
        wordCount = 0
        for line in f:
            lineCount += 1
            line = line.strip()
            wordArray = line.split(" ")
            wordCount += len(wordArray)
            
        f.close()
        
        print("number of lines: ", lineCount)
        print("number of words: ", wordCount)
```

```python
# 실습 #8: 사칙 연산 클래스

class Calc:
    def sum(self, a, b):
        result = a + b
        print("{0:s} + {1:s} = {2:s} 입니다.".format(str(a), str(b), str(result)))
    def substraction(self, a, b):
        result = a - b
        print("{0:s} - {1:s} = {2:s} 입니다.".format(str(a), str(b), str(result)))
    def multiplication(self, a, b):
        result = a * b
        print("{0:s} * {1:s} = {2:s} 입니다.".format(str(a), str(b), str(result)))
    def division(self, a, b):
        result = a / b
        print("{0:s} - {1:s} = {2:s} 입니다.".format(str(a), str(b), str(result)))
```

```python
# 실습 #9: 계산기

class my_stack:
    def __init__(self):
        self.stack = []
    def __len__(self):
        return len(self.stack)
    def push(self, a):
        self.stack.append(a)
    def pop(self):
        if not self.stack:
            return -1
        else :
            return self.stack.pop(-1)
    def peek(self):
        if not self.stack:
            return -1
        else:
            return self.stack[-1]

def calculator():
    input_clac = input("계산식 입력: ")
    input_clac_array = input_clac.strip().split(" ")
    
    num = my_stack()
    op = my_stack()
    
    for i in range(len(input_clac_array)):
        s = input_clac_array[i]
        if s.isdigit():       
            num.push(int(s))
            op_now = op.peek()
            if op_now != -1:
                if op_now == "*" or op_now == "/" or (i == len(input_clac_array) -1):
                    # 연산
                    op_now = op.pop()
                    num2 = num.pop()
                    num1 = num.pop()
            
                    result = calculator_(op_now, num1, num2)
                    num.push(result)
                    
                    # 연산자 우선수위 때문에 '밀린' 연산이 있으면 수행
                    op_now = op.pop()
                    if(op_now != -1):
                        num2 = num.pop()
                        num1 = num.pop()
                        result = calculator_(op_now, num1, num2)
                        num.push(result)
                elif not (input_clac_array[i+1] == "*" or input_clac_array[i+1] == "/"):
                    # 연산
                    op_now = op.pop()
                    num2 = num.pop()
                    num1 = num.pop()
                    result = calculator_(op_now, num1, num2)
                    num.push(result)                
        else:
            op.push(s)
    
    print("계산식 출력:", end=" ")
    for s in input_clac_array:
        print(s, end = " ")
    print("=", num.pop())
            
def calculator_(op, num1, num2):
    if op == "+":
        return num1 + num2
    elif op == "-":
        return num1 - num2
    elif op == "*":
        return num1 * num2
    elif op == "/":
        return num1 / num2
```
