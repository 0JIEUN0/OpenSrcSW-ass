```python
# 실습 #1: 생일의 서울 지역 최고 기온과 최저 기온을 시각화
# 기온 데이터: 2010-01-01 ~ 2020-01-01 (10년치)

import csv
import matplotlib.pyplot as plt

f = open('./data/weatherData_2010-2020.csv')
data = csv.reader(f, delimiter=',')

headers = next(data)
maxTemper = []
minTemper = []

# 최고기온, 최저기온 구하기
for row in data:
    month = row[0].split('-')[1]
    day = row[0].split('-')[2]
    if month == '11' and day == '16':
        if row[-1] != '' :
            maxTemper.append(float(row[-1]))
        if row[-2] != '' :
            minTemper.append(float(row[-2]))

# 그래프 출력
plt.title('Ass #1')
plt.plot(maxTemper, color='pink', label='max')
plt.plot(minTemper, color='skyblue', label='min')
plt.legend()
plt.show()

f.close()
```

```python
# 실습 #2: 서울 지역의 1월과 8월 최고 기온 데이터를 히스토그램으로 시각화
# 기온 데이터: 2010-01-01 ~ 2020-01-01 (10년치)

import csv
import matplotlib.pyplot as plt

f = open('./data/weatherData_2010-2020.csv')
data = csv.reader(f, delimiter=',')

headers = next(data)
m01 = []
m08 = []

# 1월, 8월 최고기온 구하기
for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            m01.append(float(row[-1]))
        elif month == '08':
            m08.append(float(row[-1]))

plt.title('Ass #2')
plt.hist(m01, bins=100, color='b', label='Jan')
plt.hist(m08, bins=100, color='r', label='Aug')
plt.legend()
plt.show()

f.close()
```
