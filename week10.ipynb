```python
# 실습 2: 최대값 구하기

numbers = input("Numbers? ")
newNumbers = numbers.strip().split(" ")
newNumbers = list(map(int, newNumbers))
# max(newNumbers)

max_num = newNumbers[0]
for num in newNumbers:
    if max_num < num:
        max_num = num

print("Max = ", max_num)
```

```python
# 실습 3: 암호화/복호화

def encryption(plain_text):
    
    text_list = list(reversed(plain_text.strip().split(" ")))
    cypher = []
    for text in text_list:
        for char in text:
            tmp = (chr)(ord(char) + 1)
            cypher.append(tmp)
        cypher.append(' ')

    cypher_text = ''.join(cypher)
    return cypher_text

def decryption(cypher_text):
    text_list = list(reversed(cypher_text.strip().split(" ")))
    plain = []
    for text in text_list:
        for char in text:
            tmp = (chr)(ord(char) - 1)
            plain.append(tmp)
        plain.append(' ')

    plain_text = ''.join(plain)
    return plain_text

plain_text = input("암호화할 문자열:")
print(encryption(plain_text))

cypher_text = input("복호화할 문자열:")
print(decryption(cypher_text))
```

```python
# 실습 4: 전화번호부

def number_system(dic):
    menu = 0
    while(menu != 3):
        print("1) 이름으로 검색 2) 전화번호로 검색 3) 종료")
        menu = int(input("메뉴를 선택하세요 >> "))
        if (menu == 1):
            searchForName(dic)
        elif (menu == 2):
            searchFroNumber(dic)
        elif (menu == 3):
            print("프로그램을 종료합니다.")

def searchForName(dic):
    name = input("검색할 이름을 입력하세요 >> ")
    if(name in dic):
        print(name, "의 전화번호는 ", dic.get(name), "입니다.\n")
    else:
        print("일치하는 이름이 없습니다.\n")

def searchFroNumber(dic):
    number = input("검색할 번호를 입력하세요 >> ")
    dic_keys = dic.keys()
    for key in dic_keys:
        if(number == dic.get(key)):
            print(number, "의 이름은 ", key, "입니다.\n")
            return
    print("일치하는 전화번호가 없습니다.\n")


# init
dic = {}
for i in range(10):
    name = input("이름을 입력하세요 >> ")
    number = input("전화번호를 입력하세요 >> ")
    dic[name] = number
number_system(dic) 
```
