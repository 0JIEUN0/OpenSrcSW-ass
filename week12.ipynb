```python
# 실습 #1: 서울의 일교차

# 서울에서 일교차가 가장 큰 날, 최저기온이 가장 낮은 날, 최고 기온이 가장 높은 날 출력

import csv

f = open('./data/weatherData.csv')
data = csv.reader(f, delimiter=',')

headers = next(data)

rankDate = [''] * 3
rankNum = [-999, 999, -999]

for row in data:
    if row[-1] == '':
        row[-1] = -999
    low = float(row[3])
    high = float(row[4])
    
    # 일교차
    if abs(high-low) > rankNum[0]:
        rankNum[0] = abs(high-low)
        rankDate[0] = row[0]
        
    # 최저 기온
    if low < rankNum[1]:
        rankNum[1] = low
        rankDate[1] = row[0]
        
    # 최고 기온
    if high > rankNum[2]:
        rankNum[2] = high
        rankDate[2] = row[0]


# 출력
print("일교차가 가장 큰 날")
print(rankDate[0])
print("최저기온이 가장 낮은 날")
print(rankDate[1])
print("최고기온이 가장 높은 날")
print(rankDate[2])

f.close()
```

```python
# 실습 #2: 지하철 이용객

# 노선별로 가장 이용객이 많은 3개 역과 가장 적은 3개 역을 출력

import csv

f = open('./data/subwayData.csv')
data = csv.reader(f, delimiter=',')

headers = next(data) # skip headers

stationList = []
nowStation = '1호선'
print("==============================")

for row in data:
    station = row[1]
    if station != nowStation:
        # Sorting
        stationList.sort(key=lambda x: x[1])
        
        # Print result
        print(nowStation)
        print("이용객이 가장 많은 역 순위")
        for i in range(-1, -4, -1):
            print("{0:d}. {1:s} ({2:d})".format(i*(-1), stationList[i][0], stationList[i][1]))
        print("이용객이 가장 적은 역 순위")
        for i in range(3):
            print("{0:d}. {1:s} ({2:d})".format(i+1, stationList[i][0], stationList[i][1]))
        print("==============================")
        
        nowStation = station
        stationList.clear()
    
    stationList.append((row[3],int(row[4].replace(',',''))))

f.close()
```
