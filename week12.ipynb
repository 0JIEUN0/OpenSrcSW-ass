```python
# 실습 #1: 서울의 일교차

# 서울에서 일교차가 가장 큰 날, 최저기온이 가장 낮은 날, 최고 기온이 가장 높은 날 출력

import csv

f = open('./data/weatherData.csv')
data = csv.reader(f, delimiter=',')

headers = next(data)

rankDate = [''] * 3
rankNum = [-999, 999, -999]

for row in data:
    if row[-1] == '':
        row[-1] = -999
    low = float(row[3])
    high = float(row[4])
    
    # 일교차
    if abs(high-low) > rankNum[0]:
        rankNum[0] = abs(high-low)
        rankDate[0] = row[0]
        
    # 최저 기온
    if low < rankNum[1]:
        rankNum[1] = low
        rankDate[1] = row[0]
        
    # 최고 기온
    if high > rankNum[2]:
        rankNum[2] = high
        rankDate[2] = row[0]


# 출력
print("일교차가 가장 큰 날")
print(rankDate[0])
print("최저기온이 가장 낮은 날")
print(rankDate[1])
print("최고기온이 가장 높은 날")
print(rankDate[2])

f.close()
```
